#### 2. 程序计时 `%time`和`%timeit`
* `%time`是计算中的执行时间；
* `%timeit`会进行启发式（heuristic）地一条语句跑多次，给出一个平均值；

````python
strings = ['foo', 'foobar', 'baz', 'qux', 
           'python', 'Guido Van Rossum'] * 100000
%time [x for x in strings if x.startswith('foo')]
%time [x for x in strings if x[:3] == 'foo']
````

````python
%timeit [x for x in strings if x.startswith('foo')]
%timeit [x for x in strings if x[:3] == 'foo']
````

### 一些实际的小技巧

#### 1. 通过`reload`更新模块的依赖
因为如果一些模块更新之后，重新import只会生成一个指向原来namespace的引用，没法真正进行模块的更新，需要执行reload:
```
import some_lib 
reload(some_lib)
x =5
y = [1, 2, 3, 4]
result = some_lib.get_answer(x, y)
```
如果需要进行递归更新，可以尝试`dreload(some_lib)`，但是不是所有情况下都会有效。
